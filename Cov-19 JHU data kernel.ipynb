```python
dailyfiles = sorted(glob.glob('COVID-19/csse_covid_19_data/csse_covid_19_daily_reports/*.csv'))
```

```python
dailyfiles
```

```python
df = pd.concat([pd.read_csv(file) for file in dailyfiles])
```

```python
# fill NaN
df = df.fillna({"Province/State": 'None'})

# Change astype
df["Last Update"] = pd.to_datetime(df["Last Update"])
df["Confirmed"] = df["Confirmed"].fillna(0).astype('int')
df["Deaths"] = df["Deaths"].fillna(0).astype('int')
df["Recovered"] =df["Recovered"].fillna(0).astype('int')
```
